```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import gensim
```

```python
%%time
wv = Word2Vec.load("w2v_300.model")
```

```python
variance = []
X = wv.wv.vectors
pca = PCA(n_components=300)
pca.fit(X)
variance = pca.explained_variance_ratio_
plt.figure(figsize=(10, 10))

accumulated_variance = [0]

for var in variance:
    accumulated_variance.append(accumulated_variance[-1] + var)
accumulated_variance.remove(0)

plt.bar(np.arange(len(accumulated_variance)), accumulated_variance, label='Accumulated Variance')
plt.axhline(y=0.99, color='r', linestyle='--', label='99%')

plt.title('w2v, 300, train+test')
plt.xlabel('Component Number')
plt.ylabel('Accumulated Variance')
plt.grid()
plt.legend(loc=4)

plt.show()
```
